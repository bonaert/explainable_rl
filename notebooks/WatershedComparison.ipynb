```python
import pickle
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
with open('watershed_pso.pickled', 'rb') as pso_file, open('watershed_rl.pickled', 'rb') as rl_file:
    solution_pso = pickle.load(pso_file)
    solution_rl = pickle.load(rl_file)
```

```python
for i in range(4):
    action_pso = [a[i] for (s, a) in solution_pso.values()]
    action_rl = [a[i] for (s, a) in solution_rl.values()]
    plt.figure()
    plt.boxplot(action_pso, positions=[1], labels=["PSO"])
    plt.boxplot(action_rl, positions=[2], labels=["RL"])
    
```

```python
for i in range(150):
    score_pso, vars_pso = solution_pso[i]
    score_rl, vars_rl = solution_rl[i]
    vars_pso, vars_rl = [round(x, 1) for x in vars_pso], [round(x, 1) for x in vars_rl] 
    
    print(f"PSO - {score_pso:.1f} - {vars_pso}\n"
          f"RL  - {score_rl:.1f} - {vars_rl}\n")
```
